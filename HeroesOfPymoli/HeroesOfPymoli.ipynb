```python
# import dependencies, assign main file to a variable
import pandas as pd
import numpy as np

file = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(file)
purchase_data
```

```python
# Displays total number of unique players
totalplayers = len(purchase_data["SN"].unique())

# Put into a data frame
unique_players = [{"Total Number of Unique Players": totalplayers}]
unique_players_df = pd.DataFrame(unique_players)
unique_players_df
```

```python
# Displays number of unique items, average purchase price, total number of purchases, and total revenue; puts into a data frame
uniqueitems = len(purchase_data["Item Name"].unique())
totalpurchases = len(purchase_data["Price"])
sumpurchases = purchase_data["Price"].sum()
formatsumpurchases = ('${:.2f}'.format(sumpurchases))
averagepurchase = ('${:.2f}'.format(sumpurchases/totalpurchases))

#Puts into a data frame
purchase_data_info = [{"Number of Unique Items": uniqueitems, "Average Purchase Price": averagepurchase,
                                   "Total Number of Purchases": totalpurchases, "Total Revenue": formatsumpurchases}]
purchase_data_info_df = pd.DataFrame(purchase_data_info)
purchase_data_info_df
```

```python
# Displays total count of unique individuals based on gender, and their respective percentages of the user base
gendergrouped_df = purchase_data.groupby("Gender")
unique_players = gendergrouped_df["SN"].nunique()
percentplayers = round((unique_players/totalplayers)*100, 2)

#Puts into a data frame with proper formatting
gender_info = pd.DataFrame({"Total Count": unique_players, "Percentage of Players": percentplayers})
gender_info["Percentage of Players"] = gender_info["Percentage of Players"].map("{:.2f}%".format)
gender_info
```

```python
# Displays purchase count, average purchase price, total purchase value, and average purchase total per person, based on gender
gender_counts = gendergrouped_df["Gender"].count()
average_purchase = gendergrouped_df["Price"].mean()
total_purchase_value = gendergrouped_df["Price"].sum()
average_total_purchase = total_purchase_value/unique_players

#Puts into a data frame and maps proper formatting
gender_info = pd.DataFrame({"Purchase Count": gender_counts, "Average Purchase Price": average_purchase,
                            "Total Purchase Value": total_purchase_value, "Average Purchase Total per Person": 
                             average_total_purchase})
gender_info["Average Purchase Price"] = gender_info["Average Purchase Price"].map("${:.2f}".format)
gender_info["Total Purchase Value"] = gender_info["Total Purchase Value"].map("${:.2f}".format)
gender_info["Average Purchase Total per Person"] = gender_info["Average Purchase Total per Person"].map("${:.2f}".format)

gender_info
```

```python
# Displays purchase count, average purchase price, total purchase value, and average purchase total per person, based on
# implemented age groups via bins
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=groups)
age_group = purchase_data.groupby("Age Group")
unique_players = age_group["SN"].nunique()
groups_purchase_counts = age_group["Age"].count()
groups_average_purchase = age_group["Price"].mean()
groups_total_purchase_value = age_group["Price"].sum()
groups_average_total_purchase = groups_total_purchase_value/unique_players

#Puts into a data frame and maps proper formatting
groups_info = pd.DataFrame({"Purchase Count": groups_purchase_counts, "Average Purchase Price": groups_average_purchase,
                            "Total Purchase Value": groups_total_purchase_value, "Average Purchase Total per Person": 
                             groups_average_total_purchase})
groups_info["Average Purchase Price"] = groups_info["Average Purchase Price"].map("${:.2f}".format)
groups_info["Total Purchase Value"] = groups_info["Total Purchase Value"].map("${:.2f}".format)
groups_info["Average Purchase Total per Person"] = groups_info["Average Purchase Total per Person"].map("${:.2f}".format)
groups_info
```

```python
# Displays purchase count, average purchase price, and total purchase value based on the top 5 spenders
user_group = purchase_data.groupby("SN")
user_purchase_count = user_group["Age"].count()
user_average_purchase = user_group["Price"].mean()
user_total_purchase = user_group["Price"].sum()

#Puts into a data frame and maps proper formatting
user_group_info = pd.DataFrame({"Purchase Count": user_purchase_count, "Average Purchase Price": user_average_purchase,
                                "Total Purchase Value": user_total_purchase})
user_group_info = user_group_info.sort_values("Total Purchase Value", ascending = False) #Displays in descending order
user_group_info["Average Purchase Price"] = user_group_info["Average Purchase Price"].map("${:.2f}".format)
user_group_info["Total Purchase Value"] = user_group_info["Total Purchase Value"].map("${:.2f}".format)
user_group_info.head(5)
```

```python
# Displays purchase count, average purchase price, and total purchase value based on the most popular items
popular = purchase_data.loc[:,["Item ID","Item Name","Price"]]
item_group = purchase_data.groupby(["Item ID", "Item Name"])
item_purchase_count = item_group["Age"].count()
item_total_purchase = item_group["Price"].sum()
average_purchase = item_group["Price"].mean()

#Puts into a data frame and maps proper formatting
item_info = pd.DataFrame({"Purchase Count": item_purchase_count, "Average Purchase Price": average_purchase, 
                          "Total Purchase Value": item_total_purchase})
item_info["Average Purchase Price"] = item_info["Average Purchase Price"].map("${:.2f}".format)
item_info["Total Purchase Value"] = item_info["Total Purchase Value"].map("${:.2f}".format)
item_info = item_info.sort_values("Purchase Count", ascending = False) #Displays in descending order
item_info.head(5)
```
